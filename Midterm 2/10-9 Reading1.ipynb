# Address spaces

## Early Systems

Early machines didn't provide much of an abstraction to users. The physical memory of the machine consisted of the OS and the current program.  
The OS was a set of routines that sat in memory and there would be one running process that currently sat in physical memory and used the rest of the memory.

## Multiprogramming and Time Sharing

Multiprogramming made people realize that the current approach was too slow

## The Address Space

We need to keep an abstraction of physical memory, which is called the address space, and it is the running program's view of memory.  
The address space of a process contains all of the memory state of the running program. It contains the code, the stack, and the heap of a currently running program.   
Code is static, so we place it at the top of our address space. We have two regions of the address space that may grow and shrink while the program runs. These are the heap, at the top of the program, and the stack, which is at the bottom.  
The OS virtualizes memory because the running program thinks it is loaded into memory at a particular address and has a large address space, but the reality is different. It is instead lodaded into a virtual address of 0, which is mapped to a physical memory address somewhere else.

## Goals

One major goal of VM is efficienct. The OS should strive to make virtualization as efficient as possible. Another is transparency, the OS should implement VM in a way such that it is invisible to the running program. The OS should do all the work to the memory among many different jobs. The final goal is protection. The OS should make sure to protect processes from one another as well as the OS itself from the processes. It enables us to deliver on the property of isolation between processes.

# Address Translation

LDE- Limited direct execution, let the program run directly on the hardware.   
We can add address translation, which will allow the hardware to transform each memory access, hanging the virtual address provided by the instruction to a physical address where the information is actually located. The OS must be involved at key points so that correct translation can take plact, and it must manage memory.

## Assumptions

We will assume that the user's address space must be placed contiguously in physical memory. We will also assume that the size of the address space is less than the size of the physical memory, and that each address space is the same size.

## Dynamic Relocation

The first iteration is dynamic relocation, and we need two hardware registers, one called the base and bounds and this pair will allow us to place the address space anywhere we'd like in the physical memory.  
Each program is written and compiled as if it is loaded at 0. The physical address is translated as equivalent to the virtual address + base. Each reference generated by the process is a virtual address, and the hardware adds the base register to this address, which will result in a physical address.  

## Hardware Support

The hardware must provide a base and bounds registers, so the CPU has an additional pair of registers as a part of the MMU of the CPU. When a user program is running, the hardware will translate each address.

## Operating System Issues

The OS must take action when a process is created, finding space for its address space in memory. When a new process is created, the OS will have to seach a data structure to find room for the new address space and then mark it used. The OS must do work when a process is terminated. It has to put the memory back in the free list. The OS must perform a few additional steps when a context switch occurs, so the OS must save and restore the base and bounds pair when it switches between processes.
