# Disks

Just like memory, except that memory is volatile and disks last forever.   
Memory is smalle, disks are very big. Memory is used to for short term storage, and disks reduce what must be kept in memory. Memory is expensive, disks are cheap.

# Anatomy of a magnetic disk

Sector: contiguous collection of bytes on disk, typically 512 bytes.  
Track: contiguous collection of sectors.  
Surface- cocentric collection of tracks.  
platter- thin disk covered in magnetic matereial, usually has 2 surfaces   
Cyclinder - set of tracks on different surfaces with same track index.   
Spindle: creates axis around which platters spin.   
Head: used to read/write data off disk.  
Arm: extends head over surface.  

Disk operations are in terms of radial coordinates used to be a disk address: (driver, surface, track, sector), now it is a logical block address converted by disk controller  
In order to read/write, wait for sector to appear under head, read/write sector as it spins by

Seek time: time to position head over track/cylender  
Maximum time is time to go from inntermost to outermost track, average is 1/3 of the way across disks.

Rotation time: time for sector to rotate under head, head starts reading as soon as it settles

Transfer time- time to move bytes from disk to memory.  
Surface transfer time: time to transfer one or more sequential sectors to/from surface after head reads/writes, much smaller than seek/rotation time.  
Host transfer time: time to transfer data between host memory and disk buffer

In a multiprogramming environment, a queue of disk I/O requests can form, requedsts are generated by the running program and handled by the disk driver, where the queue forms. The OS maximizes disk I/O throughput by minimizing head movement through disk head scheduling

# FCFS

Assume a queue of requests exists to read/write track and go to first request that comes in

# SSTF

Shortest seek time first, rearrange queue to shortest seek time first order

# SCAN scheduling

optimzation is that head is reset when no more requests exist between head position and edge of disk

# C-SCAN

move head in one direction until an edge of the disk is reached and then reset to the opposite edge, optimization is that the head is reset when no more requests exist between current head position and approaching edge of the disk, works off of optimized jump to the outer edge of disk

# Performance optimization- partitioning

Disks are typically paritioned to minimizes the largest possible seek time, partition is a collectino of cylinders, a partition is a logically separate disk

To get the quickest disk response time, we must iminimize seek time and rotational latency, make disk smaller, spin disks faster, shedule disk operations to minimize head movements, lay our data on disk so that related data are on nearby disks. Place commonly used files on disk.  
Pick block size carefully: If block size is too small, we will have low transfer rate because we need to perform more seeks for same amount of data. If it is too big, we will have internal fragmentation.

# Solid States Drives

No moving parts, better random access performace, less power, uses NAND, non volatile, cheap  

# Flash Durability

Flash memory stops reliably storing a bit after many erasures, after a few years without power, after a nearby cell is read many times.   
To improve durability, error correcting codes, amangement of defective pages/erasure blocks, wear leveling- spread updates to hot logical pages to many physical pages, spares for both wear leveling and managing bad pages and blocks

# File System abstractions

Presents applications with persistent, named data.  
A file is a named collection of related information on secondary storage. Has two parts:
1. Data- what a user or application puts in the file, array of untyped bytes
2. Metadata- file attributes added and managed by the OS, includes name, type, location, size, creator, creation time, security info, modification time   
   
File operations include create, open, read, write, seek, delete   
A directory provides names for files, list of human readable names   
Hard Link: mapping from each name to a specific underlying file or directory.  
Soft Link: mapping from a file name to another file name

# Create()

Creates a new file with some meta data and a name for the file in a directory. One create, the OS allocates disk space, creates metadata for the file including name, location on disk, and all file attributes, adds an entry to the directory that contains the file

# Link()

Creates a hard link-- a new name for some underlying file.  
On link, the OS adds entry to the directory with new name, increments ocunter in the file header that tracks the number of directory entries pointing to that file, will point to the same underlying file

# Unlink()

Removes a name for a file from its directory. 
If last link, file and its resources are deleted, to delete() the OS finds directory containing the file, clears headers, frees the disk blocks used by the file and its headers, removes the entry from th directory

# open()

Creates per file data structure referred to by file descriptor, returns file descriptor to the caller    
on open(), the OS checks if the file is already open by another process. If it is not, it finds the file, copies the file descrptor into the system wide open file table, checks the protection of the file against the requester's permissions, aborts if they do not match, increments the open count, creates an entry in the process' file table pointing to the entry in the system wide file table, initializes current file pointer to the start of the file, returns the index into the process' file table

# close()

closes the file.  
On close(), the OS removes the netry for the file in the process table, decrements the open count in the system file table. If the open count == 0, removes the entry in the system wide file table.

# Other commands

read() reads size bytes from an input file position into a buffer.  
write is similar to read, except it copies the buffer to the file.  
seek() updates the file pointer.   
fsync() does not return until all data is written to persistent storage.
